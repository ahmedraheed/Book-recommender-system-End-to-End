POPULAR BASED RECOMMENDER SYSTEM

```python
rating_with_name=ratings.merge(books,on='ISBN')
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
rating_with_name=ratings.merge(books,on='ISBN')
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_rating']>=250]
```

```python
famou_books=popular_df[popular_df['num_rating']>=200].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_books=famou_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_books["Image-URL-M"][0]
```

Colaborative Filtering Based Recommender System

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
y=x[x].index
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(y)]
```

```python
z=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=z[z].index
```

```python
filtered_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=filtered_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommended(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]
    data=[]
    for i in similar_items:
        item=[]
        # print(pt.index[i[0]])
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
        

    return data
        
```

```python
recommended('You Belong To Me')
```
